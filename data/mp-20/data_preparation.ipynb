```python
import pandas as pd
from pymatgen.core import Structure
```

```python
df_train = pd.read_csv("train.csv", index_col=0)
df_val = pd.read_csv("val.csv", index_col=0)
df_test = pd.read_csv("test.csv", index_col=0)
print(len(df_train), len(df_val), len(df_test))
```

```python
# Add composition column
df_train["composition"] = df_train["cif"].apply(
    lambda x: Structure.from_str(
        x, fmt="cif"
    ).composition.reduced_composition.alphabetical_formula
)
df_val["composition"] = df_val["cif"].apply(
    lambda x: Structure.from_str(
        x, fmt="cif"
    ).composition.reduced_composition.alphabetical_formula
)
df_test["composition"] = df_test["cif"].apply(
    lambda x: Structure.from_str(
        x, fmt="cif"
    ).composition.reduced_composition.alphabetical_formula
)
```

```python
# save the data
df_train.to_csv("train.csv")
df_val.to_csv("val.csv")
df_test.to_csv("test.csv")
```
